### b) Simulação

```python

lbd = 5
tmax = 2
n = 15
Prob = (np.exp(-lbd*tmax)*(lbd*tmax)**n)/math.factorial(n)

sim = np.arange(10, 5000,10, dtype = np.int)
# Ns = []
resp = []
for nsim in sim:
    cont = 0
    for s in range(0,nsim):
        te = 0
        ne = 0
        while (te < tmax):
            te = te + np.random.exponential(1/lbd)
            if(te < tmax):
                ne = ne + 1
        if(ne == n):
            cont = cont + 1
    
    resp.append(cont/nsim)
    
    
exibir(sim,resp,Prob)
    
```

```python
FIG_SIZE = (10, 6) # Tamanho para plotar usando a matplotlib.

# Essa função exibe um gráfico (mostrando os valores simulados e o valor teórico, se tiver).
def exibir(x_points, y_points, valor_teorico = None, x_lim = None, y_lim = None, sim_label = 'Valor simulado', theo_label = 'Valor teórico', x_label = 'Número de experimentos', y_label = 'Resultado obtido', figsize = FIG_SIZE):
    plt.figure(figsize = figsize)
    ax = plt.axes()
    plt.plot(x_points, y_points, linestyle='-', color = 'blue', linewidth=2, label = sim_label)
    if valor_teorico is not None:
        plt.axhline(y = valor_teorico, color = 'r', linestyle = '--', label = theo_label)
        plt.text(1.01, valor_teorico, '%g' % (valor_teorico), color = 'r', transform = ax.get_yaxis_transform(), bbox = dict(facecolor='w', alpha=0.5), ha = 'left', va = 'center')
    plt.ylabel(y_label, fontsize = 20)
    plt.xlabel(x_label, fontsize = 20)
    if x_lim is None:
        x_lim = [np.amin(x_points), np.amax(x_points)]
    if y_lim is None and valor_teorico is None:
        y_lim = [np.amin(y_points), np.amax(y_points)]
    elif y_lim is None:
        y_lim = [min(np.amin(y_points), valor_teorico), max(np.amax(y_points), valor_teorico)]
    plt.xlim(x_lim)
    plt.ylim(y_lim)
    plt.legend()
    plt.show(True)
```

```python
data = pd.read_csv('miseraveis.csv')
maxi = max(data['from'].max(),data['to'].max())
n = maxi + 1
vertices = list(range(n))
graph = nx.DiGraph()
graph.add_nodes_from(vertices)
for index, row in data.iterrows():
    graph.add_edge(row['from'], row['to'])
```

```python
plt.figure(figsize=(10,6))

options = {
    'node_color': 'pink',
    'node_size': 200,
    'width': 1,
    'arrowstyle': '-|>',
    'arrowsize': 20,
}

npos = nx.circular_layout(graph,scale=1)

nx.draw(graph,pos = npos, with_labels=True, **options, arrows=True)
plt.show()
```

```python
cura = 1
infeccoes = [10,20,30]
cont = 0
colors = ['blue','red','pink']
aux = 0

# 0 para nao infectado, 1 infectado, 2 curado
status = {} 

for i in range(n):
    status[i] = 0

tempo = 0
curr_node = 0

for infeccao in infeccoes:
    
    cont = 0

    # 0 para nao infectado, 1 infectado, 2 curado
    status = {} 

    for i in range(n):
        status[i] = 0

    tempo = 0
    curr_node = 0
    
    resp_temp = []
    resp = []


    while(cont<n):


        if(status[curr_node] == 0):
            status[curr_node] = 1


        list_for_nodes = list(graph.neighbors(curr_node))

        prob_curr_node = np.random.exponential(1/cura)

        prob_neighbors = {}

        flag = 0 


        for l in list_for_nodes:

            if(status[l] == 0):
                flag = 1
                prob_neighbors[l] = np.random.exponential(1/infeccao)


        sorted_dict = sorted(prob_neighbors.items(), key=lambda item: item[1])

        caminhou = 0

        if(flag and prob_curr_node < sorted_dict[0][1]):
            if(status[curr_node] == 1 ):
                cont+=1
                status[curr_node] = 2
                tempo += prob_curr_node

        elif(flag):
            curr_node =  sorted_dict[0][0]
            caminhou = 1
            tempo += sorted_dict[0][1]

        if(flag==0):
            if(status[curr_node] == 1):
                cont+=1
                status[curr_node] = 2
                tempo += prob_curr_node


        if(caminhou == 0):
            next_node = None
            

            for s in status:
                if(status[s]==1):
                    next_node = s
                    break
                    

            if(next_node == None):
                break

            curr_node = next_node
            

        taxa = 0
        for s in status:
            if(status[s]==1):
                taxa +=1
        
        resp.append(taxa/n) 
        resp_temp.append(tempo)



    plt.plot(resp_temp,resp,color = colors[aux],label = "infeccao = %d"%aux)
    aux +=1

plt.show(True)

        
```
